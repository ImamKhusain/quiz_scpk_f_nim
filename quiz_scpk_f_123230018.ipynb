## No 1. Operasi Matriks 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Soal 1: Perkalian Matriks A dan B
A = np.array([[8, 12], [5, 9]])  
B = np.array([[7, 9], [5, 3]])  
hasil_perkalian = np.dot(A, B)  
print(f"Hasil Perkalian Matriks A dan B:\n{hasil_perkalian}") 
```

## No 2. Operasi Matriks 2

```python
# Soal 2: Penyelesaian Sistem Persamaan Linear
x = np.array([[5, 4], [2, -1]])
y = np.array([68.8, 8.8])
nilai_xy = np.linalg.solve(x, y)
print(f"Nilai x dan y: {nilai_xy}")
```

## No 8. Data Visualization (Python GUI) 1

```python
dataCleaned = pd.read_csv("rekap_cleaned.csv")
dataCleaned.head(12)
x = dataCleaned['Bulan']
y = dataCleaned['Pengeluaran']

plt.plot(x, y, marker='o', color='r')
plt.title('Grafik Pengeluaran (januari - september)')
plt.xlabel('Bulan')
plt.ylabel('Pengeluaran')
plt.show()
```

## No 10. Data Visualization (Python GUI) 3

```python
dataCleaned = pd.read_csv("rekap_cleaned.csv")
dataCleaned.head(12)
x = dataCleaned['Bulan']
y1 = dataCleaned['Pemasukan']
y2 = dataCleaned['Pengeluaran']

plt.plot(x, y1, marker='o', color='r', label='Pemasukan')
plt.plot(x, y2, marker='o', color='b', label='Pengeluaran')
plt.legend(title = "informasi",loc= 'upper left', ncol= 1, fontsize= 'small', shadow = True, fancybox= 'True')

plt.xlabel('Bulan')
plt.ylabel('Pemasukan & Pengeluaran')
plt.show()
```
